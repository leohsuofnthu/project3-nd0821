```python
import pandas as pd
original_file = "./data/census.csv"
```

```python
df = pd.read_csv(original_file).dropna()
df.head()
```

```python
# Remove the extra spaces in column names
col_names = [col.strip() for col in df.columns]
df.columns = col_names
```

```python
df.columns
```

```python
# strip all spcace for all string in data
cat_features = [
    "workclass",
    "education",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "native-country",
]
for cat in cat_features:
    df[cat] = df[cat].str.strip()
    print(f"{cat}:\n {df[cat].unique()}")
```

```python
list(df['workclass'])
```

```python
df.to_csv('./data/cleaned_census.csv', index=False)
```
